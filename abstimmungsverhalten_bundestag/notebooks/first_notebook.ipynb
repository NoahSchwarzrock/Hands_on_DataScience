```python
import pandas as pd
import glob
import os
import regex as re
import numpy as np
```

```python
def read_all_files(path):
    data = glob.glob(path + "/*")
    
    dataframes = []
    
    for file in data:   
        df = pd.read_excel(file)
        file_name_clean =  file.split("abstimmungsdaten_xslx/")[-1].removesuffix(".xlsx")
        match = re.match(r"(\d{2}\.\d{2}\.\d{4}):(.*)", file_name_clean)
        
        if match:
            abstimmungstag = match.group(1)
            abstimmungsthema = match.group(2)
        else:
            abstimmungstag = np.nan
            abstimmungsthema = file_name_clean
        
        df["Abstimmungstag"] = abstimmungstag
        df["Abstimmungsthema"] = abstimmungsthema
        dataframes.append(df)
        
    return pd.concat(dataframes, ignore_index=True)

PATH = "../data/abstimmungsdaten_xslx"

df_laws = read_all_files(PATH)
```
