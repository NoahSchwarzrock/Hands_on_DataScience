```python
import fitz
from collections import Counter

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    font_sizes = []
    
    for page in doc:
        text_info = page.get_text("dict")
        for block in text_info["blocks"]:
            for line in block.get("lines", []):
                for span in line.get("spans", []):
                    font_sizes.append(span["size"])
    
    most_common_size = Counter(font_sizes).most_common(1)[0][0]
    
    filtered_text = []
    for page in doc:
        text_info = page.get_text("dict")
        for block in text_info["blocks"]:
            for line in block.get("lines", []):
                for span in line.get("spans", []):
                    if span["size"] == most_common_size:
                        filtered_text.append(span["text"])
    
    return " ".join(filtered_text)

extract_text_from_pdf("../data/Gebäudeenergiegesetz/2006415.pdf")
```

```python
def lemma_no_punct_no_stop(text):
    return " ".join([token.lemma_ for token in nlp(text) if not (token.is_punct or token.is_stop)]).replace("\n", "")
```

```python
import pandas as pd

# Manuelles Speichern der Dateinamen und Vektoren in einen DataFrame
df = pd.DataFrame({
    "Dateiname": [
        "geg_2006416", "geg_2006875", "geg_2006415", "geg_2006705", "geg_2007028",
        "geg_2007030", "geg_2007226", "geg_2007357", "geg_2007595", "geg_2007619",
        "geg_2007620", "geg_2007623", "geg_2007626", "geg_2008207",
        "soli_1904898", "soli_1914103", "soli_1914286", "soli_1915152", "soli_1915155"
    ],
    "Vektoren": [
        vectors_geg_2006416, vectors_geg_2006875, vectors_geg_2006415, vectors_geg_2006705, vectors_geg_2007028,
        vectors_geg_2007030, vectors_geg_2007226, vectors_geg_2007357, vectors_geg_2007595, vectors_geg_2007619,
        vectors_geg_2007620, vectors_geg_2007623, vectors_geg_2007626, vectors_geg_2008207,
        vectors_soli_1904898, vectors_soli_1914103, vectors_soli_1914286, vectors_soli_1915152, vectors_soli_1915155
    ]
})
```

```python
vector_matrix = np.vstack(df["Vektoren"])

pca = PCA(n_components=2)
pca_result = pca.fit_transform(vector_matrix)
```

```python
df["PCA_1"] = pca_result[:, 0]
df["PCA_2"] = pca_result[:, 1]
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans

num_clusters = 2
km = KMeans(n_clusters=num_clusters, random_state=42)
df["Cluster"] = km.fit_predict(df[["PCA_1", "PCA_2"]])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(x="PCA_1", y="PCA_2", hue=df["Cluster"], palette="coolwarm", data=df, edgecolor="k", alpha=0.7)

centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c="black", marker="X", s=200, label="Cluster-Zentren")

for i, row in df.iterrows():
    plt.text(row["PCA_1"], row["PCA_2"], row["Dateiname"], fontsize=9, ha='right', va='bottom')

plt.title("K-Means Clustering der Gesetzestexte (PCA-Reduktion)")
plt.xlabel("PCA Komponente 1")
plt.ylabel("PCA Komponente 2")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(df[["PCA_1", "PCA_2"]], df["Cluster"])
print(f"Silhouette Score: {silhouette_avg:.4f}")
```
